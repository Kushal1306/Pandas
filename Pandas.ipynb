# Merge Function

```python
df1=pd.DataFrame({'Roll No':[1,2,3,4,5],'Physics':[34,67,34,89,12]})
df1
```

```python
df2=pd.DataFrame({'Roll No':[1,2,3,4,5],'Chemistry':[10,20,30,40,50]})
df2
```

```python
pd.merge(df1,df2,on='Roll No')
```

```python
pd.merge(df2,df1,on='Roll No')
```

```python
pd.merge(df1,df2) #on- by7 default intersecting column is taken
```

```python
df3=pd.DataFrame({'Roll No':[1,2,3,4,5],'Physics':[34,67,34,89,12]})
df4=pd.DataFrame({'Roll No':[1,2,3,6,7],'chemistry':[34,67,34,89,12]})
pd.merge(df3,df4)
```

```python
pd.merge(df3,df4,how='left')
```

```python
pd.merge(df3,df4,how='right')
```

```python
pd.merge(df3,df4,how='outer')
```

# Append Function
Adding one dataframe into another data frame

```python
df1=pd.DataFrame({'Roll No':[1,2,3,4,5],'Physics':[34,67,34,89,12]})
df2=pd.DataFrame({'Roll No':[6,7,8,9,20],'Physics':[34,67,34,89,12]})
#df1.append(df2)
df1 = pd.concat([df1, df2], ignore_index=True)
df1
```

```python
df1=pd.DataFrame({'Roll No':[1,2,3,4,5],'Physics':[34,67,34,89,12]})
df2=pd.DataFrame({'Roll No':[6,7,8,9,20],'Chemistry':[34,67,34,89,12]})
#df1.append(df2)
df1 = pd.concat([df1, df2], ignore_index=True)
df1
```
